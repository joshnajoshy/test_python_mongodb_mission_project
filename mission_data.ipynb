```python

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
mission_df = pd.read_csv('mission_launches.csv')
mission_df.head()
```

```python
mission_df_copy = mission_df.copy()
mission_df_copy.head()
```

```python
mission_df_copy = mission_df.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
mission_df_copy.head()
```

```python
mission_df_copy['Price'] = (mission_df_copy['Price'].str.replace(",", "").astype(float))
mission_df_copy.head()
```

```python
mission_df_copy['Price'] = pd.to_numeric(mission_df_copy['Price'])
mission_df_copy['Price'] = mission_df_copy['Price'].fillna(mission_df_copy['Price'].mean().round(2))
mission_df_copy.head()
```

```python
mission_df_copy = mission_df_copy.dropna(subset=['Organisation', 'Location', 'Detail', 'Rocket_Status', 'Mission_Status'])
mission_df_copy.head()
```

```python
mission_df_copy = mission_df_copy.rename(columns={"Price": "Price (K)"})
mission_df_copy.head(15)
```

```python
# see if budget made a difference to mission success
mission_df_copy.groupby('Mission_Status')[['Price (K)']].mean()
```

```python
sns.barplot(x='Mission_Status', y='Price (K)', data=mission_df_copy)
```

```python
sns.scatterplot(x="Mission_Status", y="Price (K)", data=mission_df_copy, hue="Organisation")
```

```python
mission_df_copy.to_csv('cleaned_mission_df_copy.csv', index=False)
```
